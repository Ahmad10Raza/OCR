# Working With Images Using PIL

Pillow (PIL Fork) is a Python Imaging Library that provides extensive functionality for working with images. Here are some common operations you can perform using the Pillow library:

1. **Opening and Saving Images:**
   - Open an image: `Image.open('image.png')`
   - Save an image: `image.save('new_image.png')`

2. **Image Information:**
   - Get image size: `image.size`
   - Get image format: `image.format`

3. **Resizing and Cropping:**
   - Resize an image: `image.resize((width, height))`
   - Crop an image: `image.crop((left, top, right, bottom))`

4. **Rotating and Flipping:**
   - Rotate an image: `image.rotate(degrees)`
   - Flip an image horizontally or vertically: `image.transpose(Image.FLIP_LEFT_RIGHT)` or `image.transpose(Image.FLIP_TOP_BOTTOM)`

5. **Adding Text:**
   - Create a drawing object: `draw = ImageDraw.Draw(image)`
   - Add text to the image: `draw.text((x, y), "Hello, World", fill=(255, 0, 0))`

6. **Enhancing Image Quality:**
   - Adjust image brightness: `ImageEnhance.Brightness(image).enhance(factor)`
   - Adjust image contrast: `ImageEnhance.Contrast(image).enhance(factor)`
   - Apply sharpness filter: `image.filter(ImageFilter.SHARPEN)`

7. **Changing Color Modes:**
   - Convert to grayscale: `image.convert('L')`
   - Convert to RGB: `image.convert('RGB')`

8. **Blending and Compositing:**
   - Create a composite image: `Image.blend(image1, image2, alpha)`

9. **Drawing Shapes:**
   - Draw lines, rectangles, and ellipses using the `ImageDraw` module.

10. **Image Filtering:**
    - Apply various filters like blur, edge enhance, and more using the `ImageFilter` module.

11. **Image Concatenation:**
    - Concatenate multiple images horizontally or vertically.

12. **Image Format Conversion:**
    - Convert between different image formats (e.g., PNG to JPEG).

Here's a brief example of resizing an image with Pillow:

```python
from PIL import Image

# Open an image
image = Image.open('input.jpg')

# Resize the image to a new width and height
resized_image = image.resize((300, 200))

# Save the resized image
resized_image.save('output.jpg')
```

Pillow is a versatile library for image processing in Python, and these are just some of the basic operations you can perform. You can combine these operations to achieve more complex image processing tasks.
